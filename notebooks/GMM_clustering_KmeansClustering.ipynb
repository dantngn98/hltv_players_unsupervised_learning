```python
import polars as pl
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pickle
```

```python
with open('/Users/joshpelayo/DATA 5322 02 Statistical Machine Learning 2/Data/pca_data.pkl', 'rb') as f:
      data = pickle.load(f)
```

```python
df = data['df']
X = data['X']
X_scaled = data['X_scaled']
X_pca = data['X_pca']
pca = data['pca']
feature_cols = data['feature_cols']

X.shape, X_scaled.shape, X_pca.shape
```

```python
df.head()
```

```python
# Tried clusters from 2 to 11
ks = range(2, 11)
inertias = []
silhouettes = []

for k in ks:
    km = KMeans(n_clusters=k, random_state=0, n_init=20).fit(X_pca)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X_pca, km.labels_))

cluster_scores = pd.DataFrame({
    'k': list(ks),
    'inertia': inertias,
    'silhouette': silhouettes
})

cluster_scores
```

Overall, we can see that at k=2, the silhouette is the largest at 0.31. 

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(cluster_scores['k'], cluster_scores['inertia'], marker='o')
ax.set_xlabel('Number of clusters, K')
ax.set_ylabel('K-means inertia')
ax.set_title('Elbow Plot for K-Means');
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(cluster_scores['k'], cluster_scores['silhouette'], marker='o')
ax.set_xlabel('Number of clusters, K')
ax.set_ylabel('Silhouette score')
ax.set_title('Silhouette Scores for K-Means');
```

## Fit the final K-means model

We tried both k=2 and k=3. k=2 has a higher silhouette score, so the final cluster will be k=2

### Plot k=2

```python
kmeans = KMeans(n_clusters=2,
                random_state=0,
                n_init=20).fit(X_pca)

kmeans.labels_
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_title('K-Means Clustering Results with K=2');
```

### Plot k=3

```python
kmeans3 = KMeans(n_clusters=3,
                random_state=0,
                n_init=20).fit(X_pca)

kmeans3.labels_
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans3.labels_)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_title('K-Means Clustering Results with K=3');
```

Since k=2 is better, we will use that for our final cluster results.

```python
cluster_results = X.copy().reset_index()
cluster_results['cluster'] = kmeans.labels_

if hasattr(df, 'select'):
    name_cols = [col for col in ['player_id', 'player_name', 'player_fullname'] if col in df.columns]
    if name_cols:
        names_pd = pd.DataFrame(df.select(name_cols).to_dicts())
        cluster_results = names_pd.merge(cluster_results, on='player_id', how='right')
else:
    name_cols = [col for col in ['player_id', 'player_name', 'player_fullname'] if col in df.columns]
    if name_cols:
        cluster_results = df[name_cols].merge(cluster_results, on='player_id', how='right')

cluster_results.head(10)
```
